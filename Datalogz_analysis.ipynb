```python
import pandas as pd
```

```python
# Load the CSV file
file_path = "Reports_Metric_Table_Demo.csv"  # Update the path accordingly
df = pd.read_csv(file_path)
```

```python
# Display basic info
df.head()  # Show the first few rows
```

```python
# Checking for missing values
print("Missing Values:\n", df.isnull().sum())
```

```python
# Dropping rows with critical missing values (date columns)
df = df.dropna(subset=['CREATED_DATE_TIME', 'MODIFIED_DATE_TIME'])
```

```python
# Filling other missing values with 'Unknown'
df.fillna("Unknown", inplace=True)
```

```python
# Verifying if missing values are handled
df.isnull().sum()
```

```python
# Display dataset info
print("Dataset Info:\n")
df.info()
```

```python
# Summary statistics
print("\nSummary Statistics:\n")
df.describe(include='all')  # Includes categorical columns as well
```

```python
# Convert date columns to datetime
df['CREATED_DATE_TIME'] = pd.to_datetime(df['CREATED_DATE_TIME'], errors='coerce')
df['MODIFIED_DATE_TIME'] = pd.to_datetime(df['MODIFIED_DATE_TIME'], errors='coerce')
```

```python
# Create a Year-Month column for time-based analysis
df['YearMonth'] = df['CREATED_DATE_TIME'].dt.to_period('M')
```

```python
# Check the updates
df[['CREATED_DATE_TIME', 'MODIFIED_DATE_TIME', 'YearMonth']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set a consistent style for plots
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
report_type_counts = df['REPORT_TYPE'].value_counts().head(10)

sns.barplot(x=report_type_counts.values, y=report_type_counts.index, 
            hue=report_type_counts.index, palette="viridis", legend=False)

plt.xlabel("Count")
plt.ylabel("Report Type")
plt.title("Top 10 Report Types")
plt.show()
```

```python
plt.figure(figsize=(8,4))
workspace_type_counts = df['WORKSPACE_TYPE'].value_counts()

sns.barplot(x=workspace_type_counts.values, y=workspace_type_counts.index, 
            hue=workspace_type_counts.index, palette="coolwarm", legend=False)

plt.xlabel("Count")
plt.ylabel("Workspace Type")
plt.title("Workspace Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(16,8))
reports_per_month = df['YearMonth'].value_counts().sort_index()
reports_per_month.plot(kind='line', marker='o', color='b')
plt.xlabel("Year-Month")
plt.ylabel("Number of Reports")
plt.title("Reports Created Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,4))
dedicated_capacity_counts = df['IS_ON_DEDICATED_CAPACITY'].value_counts(normalize=True) * 100
dedicated_capacity_counts.plot(kind='bar', color=['green', 'red'])
plt.ylabel("Percentage")
plt.title("Reports on Dedicated Capacity (%)")
plt.xticks([0, 1], ["No", "Yes"], rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,5))
top_workspaces = df['WORKSPACE_NAME'].value_counts().head(10)

sns.barplot(x=top_workspaces.values, y=top_workspaces.index, 
            hue=top_workspaces.index, palette="magma", legend=False)

plt.xlabel("Count")
plt.ylabel("Workspace Name")
plt.title("Top 10 Workspaces by Report Count")
plt.show()
```

```python
import numpy as np
```

```python
plt.figure(figsize=(8,5))
df['DayOfWeek'] = df['CREATED_DATE_TIME'].dt.day_name()
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

sns.countplot(x=df['DayOfWeek'], order=day_order, 
              hue=df['DayOfWeek'], palette="coolwarm", legend=False)

plt.xlabel("Day of the Week")
plt.ylabel("Number of Reports")
plt.title("Reports Created Per Day of the Week")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['Hour'] = df['CREATED_DATE_TIME'].dt.hour
sns.histplot(df['Hour'], bins=24, kde=True, color="green")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Reports")
plt.title("Reports Created Per Hour of the Day")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['TimeToModify'] = (df['MODIFIED_DATE_TIME'] - df['CREATED_DATE_TIME']).dt.total_seconds() / 3600  # Convert to hours
sns.histplot(df['TimeToModify'], bins=50, kde=True, color="purple")
plt.xlabel("Time to Modify Report (hours)")
plt.ylabel("Number of Reports")
plt.title("Distribution of Time Taken to Modify Reports")
plt.xlim(0, df['TimeToModify'].quantile(0.99))  # Remove extreme outliers
plt.show()
```

```python
pivot_table = df.pivot_table(index=df['CREATED_DATE_TIME'].dt.year, 
                             columns=df['CREATED_DATE_TIME'].dt.month, 
                             values='REPORT_NAME', 
                             aggfunc='count')

plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, cmap="Blues", annot=True, fmt=".0f")
plt.xlabel("Month")
plt.ylabel("Year")
plt.title("Heatmap of Reports Created Per Month-Year")
plt.show()
```
